## 度数分布とヒストグラム



```kotlin
%use lets-plot
%use dataframe
```

```kotlin
val classNmae = listOf(
    "0 <= point < 10",
    "10 <= point < 20",
    "20 <= point < 30",
    "30 <= point < 40",
    "40 <= point < 50",
    "50 <= point < 60",
    "60 <= point < 70",
    "70 <= point < 80",
    "80 <= point < 90",
    "90 <= point < 100",
)
val clasValue = listOf(5, 15, 25, 35, 45, 55, 65, 75, 85, 95)
val frequency = listOf(12, 10, 19, 42, 72, 82, 54, 38, 25, 19)
val relativeFrequency = frequency.map { it ->
    it.toDouble() / frequency.sum().toDouble()
}
val cumulativeFrequency =
    frequency.runningFold(0.0) { acc, x -> acc + x }.drop(1)
val cumulativeRelativeFrequency =
    relativeFrequency.runningFold(0.0) { acc, x -> acc + x }.drop(1)

val df = dataFrameOf(
    "className" to classNmae,
    "classValue" to clasValue,
    "frequency" to frequency,
    "relativeFrequency" to relativeFrequency,
    "cumulativeFrequency" to cumulativeFrequency,
    "cumulativeRelativeFrequency" to cumulativeRelativeFrequency
)

val data = df.toMap()
```

```kotlin
val plot = letsPlot(data) +
        geomBar(
            stat = Stat.identity,
            width = 1.0
        ) {
            x = "classValue"
            y = "frequency"
        } +
        labs(
            x = "Point",
            y = "Frequency",
            title = "Histogram (Frequency)"
        )

plot
```

```kotlin
val plot = letsPlot(data) +
        geomBar(
            stat = Stat.identity,
            width = 1.0
        ) {
            x = "classValue"
            y = "cumulativeFrequency"
        } +
        labs(
            x = "Point",
            y = "cumulativeFrequency",
            title = "Histogram (cumulativeFrequency)"
        )

plot
```

```kotlin
val plot = letsPlot(data) +
        geomBar(
            stat = Stat.identity,
            width = 1.0
        ) {
            x = "classValue"
            y = "cumulativeRelativeFrequency"
        } +
        labs(
            x = "Point",
            y = "cumulativeFrequency",
            title = "Histogram (cumulativeRelativeFrequency)"
        )

plot
```

## ローレンツ曲線

```kotlin
val companySize = listOf(
    "0人",
    "1-4人",
    "5-9人",
    "10-29人",
    "30-49人",
    "50-99人",
    "100-299人",
    "300-499人",
    "500-999人",
    "1000人以上"
)

val numfOfficesFreq = listOf(0.0, 4428.0, 1118.0, 705.0, 124.0, 75.0, 36.0, 4.5, 2.4, 1.1)
val relFreqOffice = numfOfficesFreq.map { it ->
    it.toDouble() / numfOfficesFreq.sum().toDouble()
}
val cumRelFreqOffice =
    relFreqOffice.runningFold(0.0) { acc, x -> acc + x }.drop(1)

val numOfEmployeesFreq = listOf(0.0, 9486.0, 7214.0, 11134.0, 4648.0, 5103.0, 5734.0, 1706.0, 1651.0, 2320.0)
val relFreqEmployees = numOfEmployeesFreq.map { it ->
    it.toDouble() / numOfEmployeesFreq.sum().toDouble()
}
val cumRelFreqEmployees =
    relFreqEmployees.runningFold(0.0) { acc, x -> acc + x }.drop(1)

println(relFreqOffice)
println(relFreqEmployees)

val dfLorenz = dataFrameOf(
    "従業者規模" to companySize,
    "事業所数" to numfOfficesFreq,
    "累積早退度数（事業所数）" to cumRelFreqOffice,
    "従業員数" to numOfEmployeesFreq,
    "累積早退度数（従業員数）" to cumRelFreqEmployees
)

dfLorenz

val dataLorenz = dfLorenz.toMap()
val dataEquality = mapOf(
    "x" to listOf(0.0, 1.0),
    "y" to listOf(0.0, 1.0)
)

val p = letsPlot(dataLorenz) {
    x = "累積早退度数（事業所数）"
    y = "累積早退度数（従業員数）"
} +
        geomLine(
            data = dataEquality,
            inheritAes = false,
            linetype = "dashed",
            color = "gray"
        ) {
            x = "x"
            y = "y"
        } +
        geomLine(size = 1.5) +
        geomPoint(size = 3) +
        scaleXContinuous(
            name = "事業所数の累積相対度数",
            limits = 0.0 to 1.0
        ) +
        scaleYContinuous(
            name = "従業員数の累積相対度数",
            limits = 0.0 to 1.0
        ) +
        coordFixed(ratio = 1.0) +
        ggtitle("Lorenz Curve（事業所規模別）") +
        themeMinimal()

p
```
